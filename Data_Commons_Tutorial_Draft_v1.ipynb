# Section 1: DataCommons

## Introduction to Data Commons

Google's Data Commons is a Knowledge Graph (KGs) that combines and organises data from multiple sources into a single database. It captures information about entities of interest in diverse domains and forges connections between them while standarizing the data into a unified schema.  An easy way to get a sense of the knowledge graph is to view it as semantic network with nodes and edges, allowing for both storage & organization of information.  


### Data Commons Knowledge Graph Structure.

Data Commons structure leverages the principles of knowledge graphs, where data is represented as nodes (entities) and edges (relationships) in a graph structure.

Each node in the graph represents an entity, such as a location, event, or statistical measure, while edges represent the relationships between these entities, like "located in" or "has population." The standard graph elements include entities, relationships, and properties. Entities are the core objects or concepts, relationships define how these entities are connected, and properties provide additional attributes or metadata about the entities, such as their values or classifications.

If you have never dealt with a graph-structure information before, imagine a graph structure as a map that shows how different things are connected. In this map, the things themselves are called "nodes" (or "vertices"). These nodes could be anything—like a person, a city, a book, or even a concept like "happiness." Then, the lines that connect these nodes are called "edges." These edges represent relationships or connections between the nodes.

For example, if you have a node for "Alice" and a node for "Bob," and they are friends, there would be an edge between these two nodes labeled "friends with." If you have another node for "New York City," and Alice lives there, there would be an edge connecting "Alice" to "New York City" labeled "lives in."


In the tutorial below we mainly use states as nodes/entities and gather information about them through properties and arcs which represent edges or relationships between different entities. The properties and entities in Data Commons graphs are inter-connected and can become one or the other. Let's demonstrate how those concepts work together in theory.



### Entities

Data Commons has a semantic graph structure where entities are nodes and their relationships are edges. An entity can represent diverse concepts, such as countries, states, cities, schools, storms, plants, statistical variables, etc. Every entity has a unique identifier, called DCID (data common's unique identifier). The DCID is the most important parameter of the API call to retrive information and data.

## Visualizing the knowledge graph

As DataCommons is knowledge graph with nodes and edges,  we can understand its network structure by using information stored as [Triples](https://docs.datacommons.org/glossary.html#triple) (three-part grouping describing node and edge objects) for any entity.

To retrieve this information, we can use the Python API method: *get_triples(dcid)*, which requieres an entity's DCID. Lets use as an example Florida,  DCID: geoId/06




### Use data commons Pyhton API to get triples

```python
import datacommons as dc
import random

# DCID for Florida
dcid = 'geoId/06'

# Get triples related to the DCID
triples = dc.get_triples([dcid])

# Print
print(f"Triples related to Florida: {len(triples[dcid])}")



```

Since the state of Florida is related to 5790 nodes, lets select 15 random nodes to visualize them.

```python
#Choose ten random triples, change the random seed number to get new random results
random.seed(68)
triples=random.sample(triples[dcid], 15)

#Print the triples
for triple in triples:
    print(triple)
```

### Lets use Network X to visualize these nodes.

```python
import networkx as nx
import matplotlib.pyplot as plt
#Visualize the triples using Networkx
plt.figure(figsize=(6, 6))
G = nx.DiGraph()

# Add edges to the graph
for s, p, o in triples:
    G.add_edge(s, o, label=p)

# Draw the graph
pos = nx.circular_layout(G)
labels = nx.get_edge_attributes(G, 'label')
nx.draw(G, pos, with_labels=True, node_size=500, node_color="orange", font_size=8)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color='darkblue',font_size=8)

plt.show()
```

# Section 2: Emissions data

## Getting the data

To understad the potential and use of Data Commons for data, we are going to visualize the main greenhouse gas emissions in the United States per state. For this we are using the Python API and we are going to need the DCIDs of the variables of interest:


- 'Count_Person',
- 'Annual_Emissions_CarbonDioxide_Biogenic',
- 'Annual_Emissions_NitrousOxide_NonBiogenic',
- 'Annual_Emissions_Methane_NonBiogenic',
- 'Annual_Emissions_CarbonDioxide_NonBiogenic'

Here's the step-by-step guide on how to get this data for every state in the US for some time period.

1. *First*, we start off by importing the DataCommons Python-wrapped API libraries.

```python
import datacommons as dc
import datacommons_pandas
import pandas as pd
```

```python
## Get all the states DCIDs with the get_places_in method

states = dc.get_places_in(['country/USA'],place_type='State') # returns a dictionary

states_list = states['country/USA']
```

```python
## list statistical variables for each state
variables_list = [
    'Count_Person',
    'Annual_Emissions_CarbonDioxide_Biogenic',
    'Annual_Emissions_NitrousOxide_NonBiogenic',
    'Annual_Emissions_Methane_NonBiogenic',
    'Annual_Emissions_CarbonDioxide_NonBiogenic'
]
```

```python
# For each state, get additional information using the get_property_values method.
# name, longitude, latitude, and nearby places (list of states and distance to original state )

names = dc.get_property_values(states_list, 'name')
nearby_places = dc.get_property_values(states_list, 'nearbyPlaces')
lat = dc.get_property_values(states_list, 'latitude')
lon = dc.get_property_values(states_list, 'longitude')
```

```python
## Use the get_stats_series method to get the timeseries data for the variables we selected for each state
data = []

for state in states_list:
    for var in variables_list:
        try:
            stat_series = dc.get_stat_series(state, var)
            for year, value in stat_series.items():
                data.append({'State': state, 'Year': year, 'Variable': var, 'Value': value})
        except Exception as e:
            print(f"Error fetching data for {state} and {var}: {e}")
```

```python
# Transform the list of results into a dataframe
df = pd.DataFrame(data)

# Pivot the dataframe to have a tidy dataset in wide  format
df = df.pivot_table(index=['State', 'Year'], columns='Variable', values='Value').reset_index()
```

```python
## Add static information
df['Name'] = df['State'].map(names)
df['Nearby_Places'] = df['State'].map(nearby_places)
df['Lat'] = df['State'].map(lat)
df['Lon'] = df['State'].map(lon)
```

```python
df
```

```python
# calculate emissions per capita and delete observations without with emissions data

df = df[pd.notnull(df['Annual_Emissions_CarbonDioxide_NonBiogenic'])]

df['CO2_NB_per_capita'] = df['Annual_Emissions_CarbonDioxide_NonBiogenic'] / df['Count_Person']
df['N04_per_capita'] = df['Annual_Emissions_NitrousOxide_NonBiogenic'] / df['Count_Person']
df['CH4_per_capita'] = df['Annual_Emissions_Methane_NonBiogenic'] / df['Count_Person']
df['CO2_B_per_capita'] = df['Annual_Emissions_CarbonDioxide_Biogenic'] / df['Count_Person']
```

```python
# clean Name variable
df['Name'] = df['Name'].apply(lambda x: x[0])

#clean Nearby places variable

def get_n_dcids(neighbors_list, n=5):
    sorted_nearby = sorted(neighbors_list, key=lambda geo: float(geo.split("@")[1][:-1]))
    top_n_dcids = [geo.split("@")[0] for geo in sorted_nearby[:n]]
    return top_n_dcids

df['nearby_5'] = df['Nearby_Places'].apply(lambda x: get_n_dcids(x))
```



## Visualizing emissions data, static


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_emissions_nearby(my_df,state_name, var_name):
    """
    Lineplot for emissions. params are dataframe and plot variable
    """
    df_plot=my_df.copy()
    state_id=df_plot[df_plot['Name']==state_name]['State']
    state_id=state_id.unique()[0]
    ids=df_plot[df_plot['Name']==state_name]['nearby_5'].iloc[0]
    ids.append(state_id)
    df_plot=df_plot[df_plot['State'].isin(ids)]

    df_plot["Year"] = pd.to_datetime(df_plot["Year"], format="%Y")
    df_plot['alpha'] = np.where(df_plot['Name']==state_name, 1.0, 0.5)


    plt.figure(figsize=(12, 8))

    for state in df_plot['Name'].unique():
        state_data = df_plot[df_plot['Name'] == state]
        sns.lineplot(
            data=state_data, x="Year", y=var_name, hue="Name", palette="tab10",
            linewidth=2.5, alpha=state_data['alpha'].iloc[0]
        )
        last_year = state_data['Year'].max()
        last_value = state_data[state_data['Year'] == last_year][var_name].values[0]
        plt.text(last_year, last_value, state, horizontalalignment='left', size='medium', color='black', weight='semibold')

    handles, labels = plt.gca().get_legend_handles_labels()
    new_labels = [
        f"{state}: {dcid}"
        for state, dcid in zip(df_plot["Name"].unique(), df_plot["State"].unique())
    ]

    plt.legend(handles=handles, labels=new_labels, title="State: DCID",
               loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)

    plt.title(f"{var_name}, {state_name} and nearby states")
    plt.xlabel("Year")
    plt.ylabel(f"{var_name}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Sample DataFrame
# data = {
#     'State': ['geoId/01', 'geoId/02', 'geoId/03', 'geoId/04', 'geoId/05', 'geoId/06'],
#     'Year': [2020, 2020, 2020, 2020, 2021, 2020],
#     'Annual_Emissions_CarbonDioxide_Biogenic': [100, 200, 150, 300, 250, 400],
#     'Annual_Emissions_CarbonDioxide_NonBiogenic': [90, 180, 140, 290, 230, 380],
#     'Count_Person': [5000000, 700000, 1000000, 400000, 3000000, 2500000],
#     'Nearby_Places': [['geoId/02', 'geoId/03', 'geoId/04', 'geoId/05', 'geoId/06'],
#                       ['geoId/01', 'geoId/03', 'geoId/04'],
#                       ['geoId/01', 'geoId/02', 'geoId/04', 'geoId/05', 'geoId/06'],
#                       ['geoId/01', 'geoId/02', 'geoId/03', 'geoId/05', 'geoId/06'],
#                       ['geoId/01', 'geoId/02', 'geoId/03', 'geoId/04', 'geoId/06'],
#                       ['geoId/01', 'geoId/02', 'geoId/03', 'geoId/04', 'geoId/05']],
#     'Name': ['[alabama]', '[alaska]', '[arizona]', '[arkansas]', '[california]', '[colorado]']
# }
data_plot  = df
÷# df = df_pivot[df_pivot['Year'] >= 2016]

# df['Nearby_Places'] = df['Nearby_Places'].apply(lambda k: [x.split('@')[0] for x in k])

# Function to create the interactive plot
def create_plotly_visualization(df):
    states = df['State'].unique()
    years = df['Year'].unique()

    # Initialize figure
    fig = go.Figure()

    # Add dropdowns
    fig.update_layout(
        updatemenus=[
            {
                'buttons': [
                    {
                        'method': 'update',
                        'label': state,
                        'args': [
                            {'visible': [state in trace.name for trace in fig.data]},
                            {'title': f"State: {state}"}
                        ]
                    } for state in states
                ],
                'direction': 'down',
                'showactive': True,
                'x': 0.1,
                'xanchor': 'left',
                'y': 1.1,
                'yanchor': 'top'
            },
            {
                'buttons': [
                    {
                        'method': 'update',
                        'label': str(year),
                        'args': [
                            {'visible': [str(year) in trace.name for trace in fig.data]},
                            {'title': f"Year: {year}"}
                        ]
                    } for year in years
                ],
                'direction': 'down',
                'showactive': True,
                'x': 0.3,
                'xanchor': 'left',
                'y': 1.1,
                'yanchor': 'top'
            }
        ]
    )

    # Add traces for each state and year
    for state in states:
        for year in years:
            filtered_df = df[(df['State'] == state) & (df['Year'] == year)]
            if not filtered_df.empty:
                nearby_places = filtered_df['Nearby_Places'].values[0]
                nearby_df = df[(df['State'].isin(nearby_places)) & (df['Year'] == year)]
                combined_df = pd.concat([filtered_df, nearby_df])

                fig.add_trace(
                    go.Bar(
                        x=combined_df['Name'],
                        y=combined_df['CO2_NB_per_capita'],
                        name=f"{state} {year} Co2 NB Biogenic per capita",
                        visible=False
                    )
                )

                fig.add_trace(
                    go.Bar(
                        x=combined_df['Name'],
                        y=combined_df['N04_per_capita'],
                        name=f"{state} {year} N04_per_capita",
                        visible=False
                    )
                )

                fig.add_trace(
                    go.Bar(
                        x=combined_df['Name'],
                        y=combined_df['CH4_per_capita'],
                        name=f"{state} {year} CH4_per_capita",
                        visible=False
                    )
                )

    # Set the initial visibility
    initial_state = states[0]
    initial_year = years[0]
    fig.update_traces(visible=False)
    for trace in fig.data:
        if initial_state in trace.name and str(initial_year) in trace.name:
            trace.visible = True

    # Set layout
    fig.update_layout(
        title=f"State: {initial_state}, Year: {initial_year}",
        xaxis_title="State",
        yaxis_title="Value",
        barmode='group'
    )

    return fig

# Create the plot
fig = create_plotly_visualization(df)

# Show the plot
fig.show()
```

# Section 3: Traversing Data Commons knowledge graph
Section 2 shows how Data Commons facilitates the retrieval of statistical information for specific entities. However, the knowledge graph stores a lot more information and datasets that can be accessed using different methods of the Python API.

Recalling the example of the introduction, we can explore how nodes are connected to each other using the Triples method: **dc.get_triples([dcid])**. However, we can use other methods to traverse different levels and types of information:

## dc.get_places_in
With **dc.get_places_in(dcids, place_type)** you can retrieve place contained in other places. (Note that containedInPlace is also a property of certain entities). Place types include globally and partialy available adminstritave divisions, US specific adminstrative divisions, European divisions, and much [more](https://docs.datacommons.org/place_types.html)

Run the following code to visualize how get_places_in lets us traverse the knowledge graph. We will use a top-down approach for US specific administrative areas.


```python
# Lets start with a top down approach for US specefic administrative areas.
# Start with USA and step to other administrative divisions

# Get states in USA:
states = dc.get_places_in(['country/USA'],place_type='State')['country/USA']

# For each state, get their counties:
counties = dc.get_places_in(states, place_type='County')

all_counties = []
for lists in counties.values():
    all_counties.extend(lists)

# For each county, get their Zip Codes:
zips = dc.get_places_in(all_counties, place_type='CensusZipCodeTabulationArea')

```

```python

# Build a hierarchical network with Networkx
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Level 1: USA
root = "USA"
G.add_node(root)
# Level 2:States, select only three for clarity in the graph
first_three_states = states[:3]
G.add_nodes_from(first_three_states)

# Level 3: Counties
reduced_counties= {state: counties[state] for state in first_three_states if state in counties}

G.add_nodes_from(reduced_counties)

# Add edges

for node in first_three_states:
    G.add_edge(root, node)

for parent, children in reduced_counties.items():
    for child in children:
        G.add_edge(parent, child)

# Omit zip codes for clarity in the graph


def custom_layout(G):
    pos = {}
    pos[root] = (0.5, 1)
    for i, node in enumerate(first_three_states):
        pos[node] = (i / (len(first_three_states) - 1), 0.75)
    for i, (parent, children) in enumerate(reduced_counties.items()):
        for j, child in enumerate(children):
            pos[child] = (i / (len(reduced_counties) - 1), 0.5 - j * 0.1)
    return pos

# Draw the graph
pos = custom_layout(G)
nx.draw_networkx_nodes(G, pos, node_color="skyblue", node_size=500)
nx.draw_networkx_labels(G, pos, font_size=8, font_weight="bold")
nx.draw_networkx_edges(G, pos, edge_color="black", arrows=True, alpha=0.5)

plt.title("")
plt.show()
```
