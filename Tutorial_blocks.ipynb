# Section 1: DataCommons


## Introduction to Data Commons

Google's Data Commons is a Knowledge Graph (KGs) that combines and organises data from multiple sources into a single database. It captures information about entities of interest in diverse domains and forges connections between them while standarizing the data into a unified schema.  An easy way to get a sense of the knowledge graph is to view it as semantic network with nodes and edges, allowing for both storage & organization of information.  


### Entities

Data Commons has a semantic graph structure where entities are nodes and their relationships are edges. An entity can represent diverse concepts, such as countries, states, cities, schools, storms, plants, statistical variables, etc. Every entity has a unique identifier, called DCID (data common's unique identifier). The DCID is the most important parameter of the API call to retrive information and data.

# Section 2: Emissions data
In this part of the tutorial we will show how to use the APIs to gather the data of interest and manipulate it.

## Getting the data

To understad the potential and use of Data Commons for data, we are going to visualize the main greenhouse gas emissions in the United States per state. For this we are using the Python API and we are going to need the DCIDs of the variables of interest:


- 'Count_Person',
- 'Annual_Emissions_CarbonDioxide_Biogenic',
- 'Annual_Emissions_NitrousOxide_NonBiogenic',
- 'Annual_Emissions_Methane_NonBiogenic',
- 'Annual_Emissions_CarbonDioxide_NonBiogenic'

````python
import datacommons as dc
import datacommons_pandas
import pandas as pd
````

````python
## Get all the states DCIDs with the get_places_in method

states = dc.get_places_in(['country/USA'],place_type='State') # returns a dictionary

states_list = states['country/USA']
````

````python
## list statistical variables for each state
variables_list = [
    'Count_Person',
    'Annual_Emissions_CarbonDioxide_Biogenic',
    'Annual_Emissions_NitrousOxide_NonBiogenic',
    'Annual_Emissions_Methane_NonBiogenic',
    'Annual_Emissions_CarbonDioxide_NonBiogenic'
]
````

````python
# For each state, get additional information using the get_property_values method.
# name, longitude, latitude, and nearby places (list of states and distance to original state )

names = dc.get_property_values(states_list, 'name')
nearby_places = dc.get_property_values(states_list, 'nearbyPlaces')
lat = dc.get_property_values(states_list, 'latitude')
lon = dc.get_property_values(states_list, 'longitude')
````

````python
## Use the get_stats_series method to get the timeseries data for the variables we selected for each state
data = []

for state in states_list:
    for var in variables_list:
        try:
            stat_series = dc.get_stat_series(state, var)
            for year, value in stat_series.items():
                data.append({'State': state, 'Year': year, 'Variable': var, 'Value': value})
        except Exception as e:
            print(f"Error fetching data for {state} and {var}: {e}")
````

````python
# Transform the list of results into a dataframe
df = pd.DataFrame(data)

# Pivot the dataframe to have a tidy dataset in wide  format
df = df.pivot_table(index=['State', 'Year'], columns='Variable', values='Value').reset_index()
````

````python
## Add static information
df['Name'] = df['State'].map(names)
df['Nearby_Places'] = df['State'].map(nearby_places)
df['Lat'] = df['State'].map(lat)
df['Lon'] = df['State'].map(lon)
````

This what the dataframe is composed of - for each state-year pair we collected all the necessary variables.

````python
df.head()
````

Now we will calculate some calculated fields for our analysis such as emissions per capita utilizing the variables we got from the API calls.

````python
# calculate emissions per capita and delete observations without with emissions data

df = df[pd.notnull(df['Annual_Emissions_CarbonDioxide_NonBiogenic'])]

df['CO2_NB_per_capita'] = df['Annual_Emissions_CarbonDioxide_NonBiogenic'] / df['Count_Person']
df['N04_per_capita'] = df['Annual_Emissions_NitrousOxide_NonBiogenic'] / df['Count_Person']
df['CH4_per_capita'] = df['Annual_Emissions_Methane_NonBiogenic'] / df['Count_Person']
df['CO2_B_per_capita'] = df['Annual_Emissions_CarbonDioxide_Biogenic'] / df['Count_Person']
````

````python
# clean Name variable
df['Name'] = df['Name'].apply(lambda x: x[0])

#clean Nearby places variable

def get_n_dcids(neighbors_list, n=5):
    sorted_nearby = sorted(neighbors_list, key=lambda geo: float(geo.split("@")[1][:-1]))
    top_n_dcids = [geo.split("@")[0] for geo in sorted_nearby[:n]]
    return top_n_dcids

df['nearby_5'] = df['Nearby_Places'].apply(lambda x: get_n_dcids(x))
````

## Visualizing emissions data, static

Now we will use cleaned and transformed data to visualize emissions.

Function Overview: plot_emissions_nearby
The plot_emissions_nearby function creates a line plot to visualize an emission variable over time for a selected state and its nearby states. Nearby states is a property we acquired on the previous step - the property of the state nodes

* Data Preparation: The function copies the input DataFrame, identifies the state and nearby states by their IDs, and filters the DataFrame to include only these states.

* Data Transformation: The Year column is converted to a datetime format, and transparency levels (alpha) are set to highlight the selected state over the nearby ones.

* Plotting: Using seaborn, the function generates line plots for each state. It annotates the last data point of each state with its name for easy identification.

* Legend Customization: The legend is modified to display state names alongside their IDs and is positioned below the plot.

* Final Adjustments: The function adds titles, labels, and a grid to the plot before displaying it.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_emissions_nearby(my_df,state_name, var_name):
    """
    Lineplot for emissions. params are dataframe and plot variable
    """
    df_plot=my_df.copy()
    state_id=df_plot[df_plot['Name']==state_name]['State']
    state_id=state_id.unique()[0]
    ids=df_plot[df_plot['Name']==state_name]['nearby_5'].iloc[0]
    ids.append(state_id)
    df_plot=df_plot[df_plot['State'].isin(ids)]

    df_plot["Year"] = pd.to_datetime(df_plot["Year"], format="%Y")
    df_plot['alpha'] = np.where(df_plot['Name']==state_name, 1.0, 0.5)


    plt.figure(figsize=(12, 8))

    for state in df_plot['Name'].unique():
        state_data = df_plot[df_plot['Name'] == state]
        sns.lineplot(
            data=state_data, x="Year", y=var_name, hue="Name", palette="tab10",
            linewidth=2.5, alpha=state_data['alpha'].iloc[0]
        )
        last_year = state_data['Year'].max()
        last_value = state_data[state_data['Year'] == last_year][var_name].values[0]
        plt.text(last_year, last_value, state, horizontalalignment='left', size='medium', color='black', weight='semibold')

    handles, labels = plt.gca().get_legend_handles_labels()
    new_labels = [
        f"{state}: {dcid}"
        for state, dcid in zip(df_plot["Name"].unique(), df_plot["State"].unique())
    ]

    plt.legend(handles=handles, labels=new_labels, title="State: DCID",
               loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)

    plt.title(f"{var_name}, {state_name} and nearby states")
    plt.xlabel("Year")
    plt.ylabel(f"{var_name}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
````

````python
plot_emissions_nearby(df,"Delaware",'CO2_NB_per_capita' )
````

# Section 2. Utilizing knowdledge graphs interconnectivity.

In this part of the tutorial we will show examples on how to use interconnectivity of the knowledge graphs. The interconnectivity allows entities to serve both as properties of a node and as separate nodes connected through arcs or edges. Here is a quick overview on arcs in Data commons.

**Understanding Arcs in Data Commons**

In the context of Data Commons, "arcs" represent the relationships or connections between different entities within the knowledge graph. These relationships are crucial for understanding how different pieces of information are linked and how to navigate the complex web of knowledge.

**Retrieving Arcs using the Data Commons Python API**

The Data Commons Python API provides methods to query and retrieve information about entities and their relationships. To get arcs, you'll typically use functions that allow you to explore the connections between entities.

To do that you need to

1. Identify the entities of interest: You'll need to specify the entities (nodes) for which you want to retrieve the arcs. You can do this by using their Data Commons IDs (dcids) or by searching for them using keywords.

2. Use API functions to retrieve arcs: The Data Commons API offers various functions to explore relationships. Some commonly used ones include:

```
dc.get_property_values(dcids, properties, out=True):
```

 This function retrieves the values of specified properties for a list of entities. You can use it to get the arcs (relationships) associated with the entities.

```
dc.get_triples(dcids, out=True):
```
 This function retrieves all the triples (subject, predicate, object) associated with the given entities. It provides a comprehensive view of the relationships.



Example

Let's say you want to explore the relationships associated with the entity "California" (dcid: geoId/06).

````python
import datacommons as dc

# Get triples associated with California
triples = dc.get_triples(["geoId/06"])

# Print the triples (subject, predicate, object)
for triple in triples['geoId/06']:
    print(triple)
````

**This will output a list of triples, where each triple represents a relationship involving California. The predicate in each triple is essentially the "arc" or connection type.**

## More on Triples

We will use Triples as a key function to illustrate the interconnectivity.

The dc.get_triples function in the Data Commons Python API returns a dictionary where:

Keys: The keys in this dictionary are the DCIDs (Data Commons IDs) of the entities you provided as input to the function.

Values: The values associated with each key (DCID) are lists of triples. Each triple is itself a list containing three elements:

Subject: The DCID or name of the subject entity in the relationship.
Predicate: The property or relationship type connecting the subject and object (this is essentially the "arc" we've been discussing).
Object: The DCID, name, or value of the object entity in the relationship.

**Key Points:**

The dc.get_triples function provides a comprehensive way to explore all the direct relationships (incoming and outgoing) associated with a given set of entities.
The **"predicate"** element in each triple is crucial for understanding the type of connection between the subject and object.
You can use this output to identify shared predicates between different entities, enabling you to discover similar relationships across various parts of the knowledge graph, as we discussed in the context of finding storm events in different states.

For example, let's take this triple

```
('stormEvent/nws1024436', 'affectedPlace', 'geoId/06')
```

Breakdown:

* Subject: 'stormEvent/nws1024436'

This represents a specific storm event within the Data Commons knowledge graph. The 'nws1024436' part likely indicates some kind of identifier or code associated with this storm.

* Predicate: 'affectedPlace'

This is the relationship or connection type between the subject and object. It indicates that the storm event (subject) affected a particular place (object).

* Object: 'geoId/06'

This represents the geographical location affected by the storm. 'geoId/06' is the Data Commons ID for the state of California.
Interpretation:

In plain language, this triple tells us that the storm event identified by 'nws1024436' affected the state of California.

````python
import networkx as nx
import matplotlib.pyplot as plt

import datacommons as dc
import random

# DCID for Florida
dcid = 'geoId/06'

# Get triples related to the DCID
triples = dc.get_triples([dcid])

# Print
print(f"Triples related to Florida: {len(triples[dcid])}")


random.seed(68)
triples=random.sample(triples[dcid], 15)

#Print the triples
for triple in triples:
    print(triple)

#Visualize the triples using Networkx
plt.figure(figsize=(6, 6))
G = nx.DiGraph()

# Add edges to the graph
for s, p, o in triples:
    G.add_edge(s, o, label=p)

# Draw the graph
pos = nx.circular_layout(G)
labels = nx.get_edge_attributes(G, 'label')
nx.draw(G, pos, with_labels=True, node_size=500, node_color="orange", font_size=8)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color='darkblue',font_size=8)

plt.show()
````

The code allows you to:

Explore the types of events that have affected a specific state.
Choose a particular event type of interest.
See details (name and affected places) for a random selection of events belonging to that type.

What the Code Does:

Fetches Data: The code begins by retrieving data from the Google Data Commons, specifically looking for relationships (known as "triples") that involve the state. A triple consists of a subject, predicate, and object—for example, "Virginia" (subject) "affectedPlace" (predicate) "CycloneEvent" (object). These triples represent facts about how the state is related to other entities.

Identifies the State: The state (Virginia) is identified by its unique identifier, and its name is fetched to use as a label in the graph. This state becomes the central node of the graph, from which all other relationships will branch out.

Focuses on Events: The code specifically looks for events that have affected the state, which are connected to the state through the "affectedPlace" relationship. It counts how many times different types of events (like floods, fires, or cyclones) have occurred in the state.

Builds the Graph:

The state is connected to an "affectedPlace" node, which serves as a hub for different event types.
Each event type, such as "fireEvent" or "floodEvent," is connected to the "affectedPlace" node, and the graph displays how many of these events have occurred in the state.
Visualizes the Relationships: The final step involves visualizing this graph using a layout that arranges the nodes and edges (connections) in a readable format. The visualization shows the state at the center, connected to the concept of "affectedPlace," and further connected to various types of events, giving a clear picture of the relationships between the state and the events that have occurred there.

````python
import datacommons as dc
import networkx as nx
import matplotlib.pyplot as plt

def generate_initial_state_graph(state_gcid):
    """
    Generates the initial portion of the state knowledge graph showing
    the state, 'affectedPlace', and event types with their counts

    Args:
        state_gcid: The GCID of the state to analyze

    Returns:
        A NetworkX graph object representing the initial knowledge graph
    """

    G = nx.DiGraph()  # Create a directed graph

    # 1. Get triples for the state
    triples = dc.get_triples([state_gcid])

    # 2. Get state name
    state_name = dc.get_property_values([state_gcid], prop='name')[state_gcid]
    if isinstance(state_name, list):
        state_name = state_name[0]

    # Add state node
    G.add_node(state_name)

    # 3. Filter for "affectedPlace" triples and count event classes
    affected_place_triples = [triple for triple in triples[state_gcid] if triple[1] == "affectedPlace"]
    event_counts = count_event_classes(affected_place_triples)

    # Add "affectedPlace" node and event type nodes with counts
    G.add_edge(state_name, "affectedPlace")
    for event_type, count in event_counts.items():
        if not isinstance(event_type, (str, int, float)):
            print(f"Warning: Unexpected type for event_type: {type(event_type)}, value: {event_type}")
            continue
        G.add_edge("affectedPlace", f"{event_type} ({count})")

    return G

# Example usage
state_graph = generate_initial_state_graph("geoId/51")  # Virginia

# Visualize the graph
pos = nx.spring_layout(state_graph)
nx.draw(state_graph, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, arrowsize=20)
plt.show()
````

## Exploration of interconnectivity tool. Now that we have a visual idea of how that works we can actually build an interconnectivity exploration tool.

The following code shows how starting with some node (in our case state-virginia) we can get to other same-level entities. So the code belows shows how to given the interest state get to know what events belong to this node and **what other entities** also belong to the shared event or arc. This showcases the interconnectivity of the structure which we will illustrate as a graph below.

The code allows you to:

Explore the types of events that have affected a specific state.
Choose a particular event type of interest.
See details (name and affected places) for a random selection of events belonging to that type.

````python
def count_event_classes(triples):
    """
    Reads triples and returns a dictionary of event classes and their counts.

    Args:
        triples: A list of triples in the format (subject, predicate, object).

    Returns:
        A dictionary where keys are event classes and values are their counts.
    """
    event_class_counts = {}
    for triple in triples:
        if triple[1] == "affectedPlace":
            event_class = triple[0].split("/")[0]  # Extract event class from subject
            event_class_counts[event_class] = event_class_counts.get(event_class, 0) + 1
    return event_class_counts
````

````python
import datacommons as dc
import random

def explore_state_events(state_dcid):
    # Step 1: Retrieve triples for the given state
    triples = dc.get_triples([state_dcid])

    # Step 2: Identify event types associated with the state
    event_types = {}
    for triple in triples[state_dcid]:
        if triple[1] == "affectedPlace":
            event_class = triple[0].split("/")[0]  # Extract event type
            if event_class not in event_types:
                event_types[event_class] = []
            event_types[event_class].append(triple[0])

    # Display the event types to the user
    print("Event types that occurred in this state:")
    for i, event_class in enumerate(event_types.keys()):
        print(f"{i + 1}: {event_class}")

    # Step 3: Allow the user to choose an event type
    chosen_type_index = int(input("Choose an event type by number: ")) - 1
    chosen_event_type = list(event_types.keys())[chosen_type_index]

    # Step 4: Ask the user how many events of this type they want to explore
    num_events = int(input(f"How many {chosen_event_type} events would you like to explore? "))

    # Get the specific events of the chosen type
    chosen_events = random.sample(event_types[chosen_event_type], num_events)

    # Step 5: For each chosen event, retrieve and display the event name and affected places
    for event_dcid in chosen_events:
        # Get the name of the event
        event_name = dc.get_property_values([event_dcid], prop='name')
        event_name = event_name[event_dcid][0] if event_name else "Unknown Event Name"

        # Get the places affected by this event
        event_triples = dc.get_triples([event_dcid])
        affected_places = [triple[2] for triple in event_triples[event_dcid] if triple[1] == "affectedPlace"]

        # Get the names of the affected places
        place_names = dc.get_property_values(affected_places, prop='name')

        # Display the event name and affected places
        print(f"\nEvent: {event_name}")
        for place_dcid, name_list in place_names.items():
            place_name = name_list[0] if name_list else "Unknown Place"
            print(f" - Affected Place: {place_name} (DCID: {place_dcid})")

# Example usage:
explore_state_events("geoId/51")
````

Graph Visualization:

Code 1: Uses default or standard settings for drawing the graph. It uses a k value of 1.5 and iterations=50 in the spring_layout, which are relatively modest settings. These parameters control the spacing between nodes and the number of iterations the layout algorithm will use to arrange the graph. This results in a more compact graph that may be harder to read if the graph becomes too dense.

Code 2: Takes extra steps to improve the readability of the graph. It increases the k value to 3, which spaces out the nodes more, and increases the number of iterations to 100, giving the layout algorithm more time to optimize the positioning of the nodes. Additionally, it reduces the node_size to 1000 (compared to 3000 in Code 1), making the nodes smaller and the graph less cluttered. Finally, it adjusts the figure size to (15, 10), creating a larger canvas for the graph.

## Code 1. Iteration with no visible relationships

````python
import datacommons as dc
import networkx as nx
import matplotlib.pyplot as plt

class StateKnowledgeGraph:
    def __init__(self, state_dcid):
        self.state_dcid = state_dcid
        self.state_name = self.get_state_name()
        self.triples = self.get_state_triples()
        self.graph = nx.DiGraph()

    def get_state_triples(self):
        """Retrieve triples for the given state."""
        return dc.get_triples([self.state_dcid])

    def get_state_name(self):
        """Retrieve the state name based on the DCID."""
        name = dc.get_property_values([self.state_dcid], prop='name')
        return name[self.state_dcid][0] if self.state_dcid in name else None

    def filter_predicates(self):
        """Filter to keep only relevant predicates."""
        relevant_predicates = {"affectedPlace", "containedInPlace", "location"}
        return [triple for triple in self.triples[self.state_dcid] if triple[1] in relevant_predicates]

    def count_event_classes(self):
        """Count event classes for affected places."""
        event_class_counts = {}
        for triple in self.triples[self.state_dcid]:
            if triple[1] == "affectedPlace":
                event_class = triple[0].split("/")[0]
                event_class_counts[event_class] = event_class_counts.get(event_class, 0) + 1
        return event_class_counts

    def filter_cyclone_triples(self):
        """Filter triples to include only those related to cyclones."""
        return [triple for triple in self.triples[self.state_dcid] if 'cyclone' in triple[0]]

    def get_cyclone_name(self, cyclone_dcid):
        """Get the name of a specific cyclone."""
        name = dc.get_property_values([cyclone_dcid], prop='name')
        return name[cyclone_dcid][0] if cyclone_dcid in name else None

    def get_affected_places(self, cyclone_dcid):
        """Get places affected by a specific cyclone."""
        triples_cyclone = dc.get_triples([cyclone_dcid])
        affected_by_cyclone = [triple for triple in triples_cyclone[cyclone_dcid] if triple[1] == "affectedPlace"]
        places_list = [triple[2] for triple in affected_by_cyclone]
        return dc.get_property_values(places_list, 'name')

    def draw_graph(self, cyclone_dcid=None):
        """Draw the knowledge graph."""
        # Add the initial state node
        self.graph.add_node(self.state_name)

        # Filter for relevant predicates
        filtered_triples = self.filter_predicates()

        # Add edges for each relevant predicate
        for triple in filtered_triples:
            self.graph.add_node(triple[2])
            self.graph.add_edge(self.state_name, triple[2], label=triple[1])

        # Add event class nodes connected to 'affectedPlace'
        event_class_counts = self.count_event_classes()
        for event_class in event_class_counts.keys():
            self.graph.add_node(event_class)
            self.graph.add_edge("affectedPlace", event_class)

        # Filter for cyclones and add cyclone-related nodes if cyclone_dcid is provided
        if cyclone_dcid:
            cyclone_name = self.get_cyclone_name(cyclone_dcid)
            self.graph.add_node(cyclone_name)
            self.graph.add_edge("cyclone", cyclone_name)

            # Add places affected by the cyclone
            affected_places = self.get_affected_places(cyclone_dcid)
            for place_dcid, place_name in affected_places.items():
                self.graph.add_node(place_name[0])
                self.graph.add_edge(cyclone_name, place_name[0])

        # Draw the graph with labels on the edges (for arcs like "affectedPlace", etc.)
        pos = nx.spring_layout(self.graph, k=1.5, iterations=50)
        nx.draw(self.graph, pos, with_labels=True, arrows=True, node_size=3000, node_color="skyblue", font_size=10)
        edge_labels = nx.get_edge_attributes(self.graph, 'label')
        nx.draw_networkx_edge_labels(self.graph, pos, edge_labels=edge_labels, font_color='red')
        plt.show()

# Example usage:
state_kg = StateKnowledgeGraph("geoId/51")
state_kg.draw_graph(cyclone_dcid='cyclone/ibtracs_2008241N19303')
````
